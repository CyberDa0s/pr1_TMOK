Функция get_config('X') возвращает набор признаков после предобработки.

Как можно увидеть, чсленные значения признаков были нормализованы с помощью Z-оценки, а категориальные призанки были закодированы с помощью метода быстрого кодирования ("one-hot encoding"). Очень важно проверить результаты нормализации, в некоторых случаях она может быть выполнена некорректно, что повлияет на возможность обучить классификатр и получить верные занчения.

Нормализация Z-оценки относится к процессу нормализации каждого значения в наборе данных, так что среднее значение всех значений равно 0, а стандартное отклонение равно 1.

Мы используем следующую формулу для нормализации z-оценки для каждого значения в наборе данных:

Новое значение = (x – μ) / σ

где:

х : Исходное значение

μ : Среднее значение данных

σ : стандартное отклонение данных

```python
# Импорт необходимых библиотек
from pycaret.classification import *  # Импортируем модуль классификации из Pycaret
import pandas as pd  # Библиотека для работы с DataFrame

# 1. Загрузка набора данных 'diabetes' из файла diabetes.txt
#data2 = pd.read_csv('/content/drive/MyDrive/Учеба/Технологии машинного обучения в кибербезопасности/1/04-task-pycaret/diabetes.txt', sep='\t')
data2 = pd.read_csv('diabetes.txt', sep='\t')

# Проверяем первые строки загруженного датасета
print(data2.head())

# 2. Настраиваем эксперимент, где целевым признаком является 'target'.
# Определяем размер тренировочного множества, задаём seed для воспроизводимости результатов
# и включаем нормализацию данных
clf = setup(
    data=data2,           # Передача исходных данных
    target='target',      # Целевой признак (переменная для предсказания)
    train_size=0.8,       # Размер тренировочной выборки (80% от всего набора данных)
    session_id=20,        # Устанавливаем фиксированный случайный seed для воспроизводимости
    normalize=True        # Включаем нормализацию признаков (приведение значений к одному масштабу)
)

# 3. Выполняем автоматический подбор лучшей модели среди предложенных моделей
# Метод compare_models() позволяет автоматически сравнивать различные модели
# и возвращает лучшую по умолчанию (по метрикам accuracy, AUC, F1 и другим).
best_model = compare_models()

# 4. Построение графика важности признаков для лучшего классификатора
# Это поможет нам увидеть, какие признаки оказывают наибольшее влияние на результат
plot_model(best_model, plot = 'feature')  # График важности признаков

# 5. Строим график кривых обучения для визуализации процесса обучения модели
# Кривые обучения помогут определить поведение модели на разных этапах тренировки
plot_model(best_model, plot = 'learning')  # График кривых обучения

# 6. Дополнительно можем сохранить выбранную модель для дальнейшего использования
save_model(best_model, 'best_diabetes_model')  # Сохраняет модель в файл

# 7. Оцениваем производительность модели на тестовом множестве
evaluate_model(best_model)  # Получаем отчеты по метрикам производительности модели

# Дополнительная возможность: создание конфьюзен матрицы для анализа ошибок классификации
plot_model(best_model, plot = 'confusion_matrix')  # Матрица путаницы (ошибок классификации)
```
